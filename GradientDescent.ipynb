
### Gradient Descent: A Mathematical Explanation

Gradient Descent is an optimization algorithm used to minimize a function by iteratively moving in the direction of steepest descent, as defined by the negative of the gradient.

#### Objective:
Minimize a loss function $ f(\theta) $, where $ \theta $ represents the parameters of the model.

#### Update Rule:
At each iteration, the parameters are updated as:
$$
\theta_{t+1} = \theta_t - \eta \nabla f(\theta_t)
$$
- $ \theta_t $: Current parameters at iteration $ t $
- $ \eta $: Learning rate (step size)
- $ \nabla f(\theta_t) $: Gradient of the loss function with respect to $ \theta_t $

#### Intuition:
- The gradient $ \nabla f(\theta_t) $ points in the direction of the steepest ascent of the function.
- Moving in the opposite direction (negative gradient) reduces the function value.
- The learning rate $ \eta $ controls the step size to ensure convergence.

#### Convergence:
Gradient Descent converges to a local minimum (or global minimum for convex functions) when:
$$
\lim_{t \to \infty} \nabla f(\theta_t) = 0
$$

#### Variants:
- **Batch Gradient Descent**: Uses the entire dataset to compute the gradient.
- **Stochastic Gradient Descent (SGD)**: Uses a single data point to compute the gradient.
- **Mini-batch Gradient Descent**: Uses a small batch of data points to compute the gradient.

Gradient Descent is widely used in machine learning to optimize model parameters by minimizing the loss function.

### Gradient Descent to Minimize $ f(x) = x^2 + 2x + 1 $

#### Function:
The function $ f(x) = x^2 + 2x + 1 $ is a convex quadratic function. It can be rewritten as:
$$
f(x) = (x + 1)^2
$$
The minimum occurs at $ x = -1 $, where $ f(x) = 0 $.

#### Gradient:
The gradient of $ f(x) $ is:
$$
\nabla f(x) = \frac{d}{dx} f(x) = 2x + 2
$$

#### Gradient Descent Update Rule:
To minimize $ f(x) $, we iteratively update $ x $ using the gradient descent rule:
$$
x_{t+1} = x_t - \eta \nabla f(x_t)
$$
where:
- $ x_t $: Current value of $ x $ at iteration $ t $
- $ \eta $: Learning rate (step size)
- $ \nabla f(x_t) = 2x_t + 2 $: Gradient at $ x_t $

#### Iterative Process:
1. Start with an initial value $ x_0 $.
2. Compute the gradient $ \nabla f(x_t) = 2x_t + 2 $.
3. Update $ x $ using $ x_{t+1} = x_t - \eta (2x_t + 2) $.
4. Repeat until convergence (i.e., $ \nabla f(x_t) \approx 0 $).

#### Convergence:
Gradient descent converges to the minimum $ x = -1 $ when the learning rate $ \eta $ is appropriately chosen. For this function, the gradient descent iteratively reduces the value of $ f(x) $ until it reaches the minimum.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the convex function and its gradient
def f(x):
    return x**2 + 2*x + 1

def grad_f(x):
    return 2*x + 2

# Gradient descent parameters
x_init = 3  # Initial point
learning_rate = 0.1
num_iterations = 20

# Perform gradient descent
x_values = [x_init]
for _ in range(num_iterations):
    x_new = x_values[-1] - learning_rate * grad_f(x_values[-1])
    x_values.append(x_new)

# Plot the function landscape
x = np.linspace(-3, 3, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = x^2 + 2x + 1", color="blue")
plt.scatter(x_values, [f(x) for x in x_values], color="red", label="Gradient Descent Iterates")
plt.plot(x_values, [f(x) for x in x_values], color="red", linestyle="--", alpha=0.7)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on a Convex Function")
plt.legend()
plt.grid()
plt.show()
```
